## 3. Number And Percentage Of Missing Values BY Column

```python
# count missing values of train data
print("Train data_summary on missing values")
pd.set_option('display.max_rows', None)
train_missing = pd.concat([df_train.isna().sum(), 100 * df_train.isna().mean()], axis=1)
train_missing.columns=['count', '%']
train_missing.sort_values(by='count')
```

```python
# count missing values of test data
print("Test data_summary on missing values")
pd.set_option('display.max_rows', None)
missing_test = pd.concat([df_test.isna().sum(), 100 * df_test.isna().mean()], axis=1)
missing_test.columns = ['count', '%']
missing_test.sort_values(by='count')
```

```python
# Extracting information into a tabular form
info_df_train = pd.DataFrame({
    'Train_Column': df_train.columns,
    'Train_NullCount': df_train.isnull().sum(),
    'Train_isnull(%)':df_train.isnull().mean() * 100,
    'Train_DataType': df_train.dtypes
})

info_df_test = pd.DataFrame({
    'Test_Column': df_test.columns,
    'Test_NullCount': df_test.isnull().sum(),
    'Test_isnull(%)':df_test.isnull().mean() * 100,
    'Test_DataType': df_test.dtypes
})

# Export to CSV
info_df_train.to_csv('Train_isnull.csv', index=False)
info_df_test.to_csv('Test_isnull.csv', index=False)
```

All six features of interest impact the sale price. Among Zoning Classes, FV(Floating Village Residential), RL(Residential Low Density) and RH(Residential High Density) are assocated with the highest prices. For Dwelling Types, the top three are 60(2-STORY 1946 & NEWER), 120(1-STORY PUD (Planned Unit Development) - 1946 & NEWER), 75(2-1/2 STORY ALL AGES). In terms of Neighborhood, NoRidge, NridgHt and StoneBr consistently have the highest porperty values. While the seasonal factor doesn't appear to significantly affect sale prices, the data shows that prices tend to peak in Sep, Nov and Dec and are lowest in June, May and April. Regarding Sale Types, New (Home just constructed and sold) and Con (Contract 15% Down payment regular terms) command higher prices than other categories. Notablly, homes classified under the Sale Condition "Partial" (Home was not completed when last assessed) have the highest sale prices, which contradicts common expectations. This finding is quite surprising ans unusual.

## 8. Derive Neighborhood Information For Market Evaluation

Based on our business knowledge, we know that factors such as neighborhood-wide supply and demand, safety, school rankings, and the envionment can significantly influence sale price. Therefore, it's essential to incorporate statistic information about each neighborhood in Ames, Iowa--such as population, area, safety, school rankings--into our dataset. 

The data includes 25 Neighborhoods in Ames, IA, but many of these names do not match the 27 neighborhoods listed on the City of Ames Neighborhood Associations Map (https://www.cityofames.org/home/showpublisheddocument/52755/637050278030900000). Using Google Maps, we observed a clear geographical pattern between neighborboods with the highest and lowest sale prices. The top five neighborhoods with the highest sale prices are located in the northwest part of Ames, while the neighborhoods with the lowest sale prices are primarily distributed in the city's central area  where the old town is situated, as well as in the southeast part of the city.
